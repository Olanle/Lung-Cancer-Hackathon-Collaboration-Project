```python
import tensorflow as tf
from django.shortcuts import render
from django.http import JsonResponse
import numpy as np

# Load the model (this should be outside of the view if it's a large model)
model = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/RAININ HACKATHON/LungCancer45666.ipynb')

def predict(request):
    if request.method == "POST":
        # Retrieve the input data, e.g., from JSON
        data = request.POST.get('data')  # Example: input data for prediction

        # Preprocess the input data (depends on your model)
        data = np.array(data).reshape(1, -1)  # Example reshape for 1D input

        # Make prediction
        prediction = model.predict(data)

        # Return prediction result
        return JsonResponse({'prediction': prediction.tolist()})
    return render(request, 'index.html')
```
